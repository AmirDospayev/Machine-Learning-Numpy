```python
import numpy as np
```

```python
nf = 3  # number of filters
rf = 3  # filter size
stride = 2
x = np.random.randn(1, 7, 7, 3)
out = np.zeros((1, 3, 3, nf))
```

```python
for k in range(x.shape[0]):
    for z in range(nf):
        h_range = int((x.shape[1] - rf) / stride) + 1
        for _h in range(h_range):
            w_range = int((x.shape[2] - rf) / stride) + 1
            for _w in range(w_range):
                out[k, _h, _w, z] = np.mean(x[k, _h * stride:_h * stride + rf, _w * stride:_w * stride + rf, :])
```

```python
# we assumed de = out
dx = np.zeros((x.shape))
```

```python
for k in range(x.shape[0]):
    for z in range(nf):
        h_range = int((x.shape[1] - rf) / stride) + 1
        for _h in range(h_range):
            w_range = int((x.shape[2] - rf) / stride) + 1
            for _w in range(w_range):
                dx[k, _h * stride:_h * stride + rf, _w * stride:_w * stride + rf, :] = out[k, _h, _w, z]
```
