```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd
sns.set()
```

```python
tesla = pd.read_csv('TSLA.csv')
tesla.head()
```

```python
plt.plot(tesla.Close)
plt.show()
```

```python
signal = np.copy(tesla.Close.values)
```

```python
def linear_weight_moving_average(signal, period):
    buffer = np.copy(signal[:period]).tolist()
    for i in range(period,len(signal)):
        buffer.append((signal[i-period:i] * (np.arange(period)+1)).sum() / (np.arange(period)+1).sum())
    return buffer
```

```python
lwma_10 = linear_weight_moving_average(signal, 10)
```

```python
plt.plot(np.arange(len(tesla.Close)), tesla.Close)
plt.plot(np.arange(len(tesla.Close)), lwma_10)
plt.show()
```
